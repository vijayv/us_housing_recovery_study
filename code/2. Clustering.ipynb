```python
import numpy as np
import pandas as pd
from sklearn import cluster
from collections import defaultdict
```

```python
master = pd.read_csv("data/master.csv")
len(np.unique(master["RegionName"]))
```

```python
df = master[master["Date"] == "2015-02"]
len(np.unique(df["RegionName"]))
```

```python
cols = ["HomesSoldAsForeclosures-Ratio", "PctOfHomesSellingForLoss", "PctOfListingsWithPriceReductions"]
pv = pd.pivot_table(df, values=cols, index="RegionName", aggfunc=np.sum)
d = pv.T.to_dict()
labels = []
data = []
for i, k in enumerate(d.keys()):
#     if i > 100:
#         break
    labels.append(k)
    data.append(d[k].values())
```

## K-Means

```python
# k = int(round(math.sqrt(len(data))))
k = 8
print k
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(data)
kmeans.cluster_centers_
kmeans.labels_
```

## Plot Clusters

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()
labels = kmeans.labels_

data = np.array(data)

ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=labels.astype(np.float))

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('HomesSoldAsForeclosures-Ratio')
ax.set_ylabel('PctOfHomesSellingForLoss')
ax.set_zlabel('PctOfListingsWithPriceReductions')
```
